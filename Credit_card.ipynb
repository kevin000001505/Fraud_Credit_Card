```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('creditcard_2023.csv')
```

```python
df.info()
```

```python
Classification = df['Class'].values
```

```python
from pca import pca
```

```python
pca()
```

```python
df.head()
```

```python
X = df.drop('Class', axis = 1).values
```

```python
model = pca(n_components = 0.85, normalize = True)
```

```python
result = model.fit_transform(X)
```

```python
#result
loadings = result['loadings']
loadings
```

```python
scores = result['PC']
scores['labels'] = Classification
scores.head()
```

```python
model.plot(figsize = (12,6))
plt.show()
```

```python
import tensorflow as tf
```

```python
# n = 12 represent at least 80%
n = 12
model = pca(n_components = n, normalize = True)
result = model.fit_transform(X)
```

```python
result2 = result['PC'].values
result2
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(result2, Classification, test_size = 0.2, random_state = 32)
```

```python
print("Training dataset : ", x_train.shape,y_train.shape )
print("Testing dataset : ", x_test.shape, y_test.shape)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
clf = MLPClassifier().fit(x_train, y_train)
```

```python
test_score = clf.score(x_test, y_test)
train_score = clf.score(x_train, y_train)
```

```python
print('train score: ', train_score)
print('test score: ', test_score)
```

### Test by cros_val

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracy = cross_val_score(clf, result2, Classification, cv = 3, scoring = 'accuracy')
print(accuracy)
print(accuracy.mean()*100,'%')
```
